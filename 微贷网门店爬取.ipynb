```python
def get_page(url,page_num,h):
    for i in range(1,page_num +1):
        formdata ={'page': i, 'rows': '10','provinceId':h}
        try:
            r = requests.post(url,data =formdata,headers=header)
            r.encoding='utf-8'
            soup=BeautifulSoup(r.text,'lxml')
            print(i)
            d=json.loads(soup.body.p.text)
            for i in range(len(d)):
                d[i]['area']=h
            x.extend(d)
            #print(i)
        except:
            print('链接异常')
            
    return x
```
